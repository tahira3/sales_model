# Amount vs Seconary factors

```python
amount=df[['online(amount)','telesale(amount)','total(amount)','AHT_per_person','connected calls','manning_count']]
amount.corr()
fig, ax = plt.subplots(figsize=(6,6))
sns.heatmap(amount.corr(),annot=True, cmap='crest', ax=ax)
plt.savefig('amount_vs_2nd_input.png')
plt.show()
```

# Count vs Seconary factors

```python
count=df[['online(count)','telesale(count)','total(count)','AHT_per_person','connected calls','manning_count']]
count.corr()
fig, ax = plt.subplots(figsize=(6,6))
sns.heatmap(count.corr(),annot=True, cmap='crest', ax=ax)
plt.savefig('count_vs_2nd_input.png')
plt.show()
```

# Average Amount vs Seconary factors

```python
avg_amount=df[['online(avg amount PU)','telesale(avg amount PU)','total(avg amount)','AHT_per_person','connected calls','manning_count']]
avg_amount.corr()
fig, ax = plt.subplots(figsize=(6,6))
sns.heatmap(avg_amount.corr(),annot=True, cmap='crest', ax=ax)
plt.savefig('avg_amount_vs_2nd_input.png')
plt.show()
```
